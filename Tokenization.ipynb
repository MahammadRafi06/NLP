```python
corpus = """Once's upon a time, in a land far away, there was a small village nestled between the mountains. 
The villagers were known for their craftsmanship, especially their ability to weave intricate tapestries. 
Each tapestry told a story, capturing the history and legends of the village.

One day, a stranger arrived, claiming to be a merchant from a distant city. He spoke of strange lands and exotic creatures, piquing the curiosity of the villagers. 
However, not everything was as it seemed. The merchant's had a secret agenda, and the villagers soon found themselves entangled in a web of deception and intrigue.
"""
```

```python
print(corpus)
```

```python
####Tokenization Techniquie1: Corpus to Sentenses

from nltk.tokenize import sent_tokenize
documnets = sent_tokenize(corpus)
```

```python
####Tokenization Techniquie1: Corpus to words or sentence to words

from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
words
```

```python
####Tokenization Techniquie1: Corpus to words or sentence to words

from nltk.tokenize import word_tokenize
words1=[]
for sentence in documnets:
    words1.append(word_tokenize(sentence))
words1
```

```python
####Tokenization Techniquie1: Corpus to words or sentence to words

from nltk.tokenize import wordpunct_tokenize ###to further divide the words that have punctuation
wordsp = wordpunct_tokenize(corpus)
wordsp
```

```python
####Tokenization Techniquie1: Corpus to words or sentence to words

from nltk.tokenize import TreebankWordTokenizer ###to count pullstops only at end 
wordst = TreebankWordTokenizer()
wordst.tokenize(corpus)
```
